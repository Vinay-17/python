```python
def count_case(string):
    upper_count = 0
    lower_count = 0
    for char in string:
        if char.isupper():
            upper_count += 1
        elif char.islower():
            lower_count += 1
    return upper_count, lower_count
string=input("Enter your string:")
upper_count, lower_count = count_case(string)
print("The number of uppercase letters is:", upper_count)
print("The number of lowercase letters is:", lower_count)
output:
Enter your string:'The quick Brow Fox'


The number of uppercase letters is: 3
The number of lowercase letters is: 12
> 
```
